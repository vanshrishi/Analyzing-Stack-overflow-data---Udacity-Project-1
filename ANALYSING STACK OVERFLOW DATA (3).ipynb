# Data Preparation

```python

#Filling NaN values in Salary column by replacing it with mean salary.
df.dropna(subset=['Salary'])
#Filling NaN values in JobSatisfaction column
js=df['JobSatisfaction'].mean()
df['JobSatisfaction'].fillna(js, inplace = True) 
js=df['CareerSatisfaction'].mean()
df['CareerSatisfaction'].fillna(js, inplace = True) 
```

I haved dropped rows with missing salary as i wanted to compute difference in salary of a professional developer who do programming as hobby , who contribute to open source projects and who have no programming hobby.Filling mean value would not provide a clear analysis instead after categorising professional developers into 3 categories(professional developer who do programming as hobby , who contribute to open source projects and who have no programming hobby) and then finding mean for particular category will help for a more clear vision on analysis.
Also where there are NaN value in Job and career Satisfaction column ,we have filled it with mean in order to take the information of that particular rows

#### ANSWER 2

```python
top_5=['United States','India','United Kingdom','Germany','Canada']
```

```python
#CREATING A DATAFRAME IN WHICH ONLY TOP 5 COUNTRIES USING STACK OVERFLOW ARE FILTERED
df1 = df[df['Country'].map(lambda x: x in top_5)].reset_index()
fig = plt.figure(figsize=(8,6))
#BOXPLOT BETWEEN JOBSATISFACTION AND COUNTRY
sns.boxplot(x="Country", y="JobSatisfaction", data=df1)
fig = plt.figure(figsize=(8,6))
#BOXPLOT BETWEEN CAREERSATISFACTION AND COUNTRY
sns.boxplot(x="Country", y="CareerSatisfaction", data=df1)
```

### CONCLUSION :
From the box plot, it is clear that Indians has less Job satisfaction as compared to other top 5 countries using stack overflow the most.The most satisfied countries in terms of job satisfaction are United States and Canada.

In terms of career satisfaction, United States and Canada are Satisfied more than any country.

#### ANSWER 3

```python
#PLOTTING PIE CHART FOR USERS , WHAT THEY OF PROGRAMING
df1['ProgramHobby'].value_counts()[0:20].plot(kind='pie',figsize=(12,10))
```

Here we get idea of distribution of 4 categories of people do programming


```python
#CREATING A FUNCTION THAT COMPUTE SALARY FOR GIVEN DATAFRAME
def Salary(dfr):
    x=dfr['Salary'].mean()
    return x
```

```python
# AVERAGE SALARY OF USER WHO HAS NO PROGRAMING HOBBY
df2=df1[df1['ProgramHobby']=='No']
NO=Salary(df2)
```

```python
# AVERAGE SALARY OF USER WHO HAS PROGRAMING HOBBY
df3=df1[df1['ProgramHobby']=='Yes, I program as a hobby']
YES=Salary(df3)
```

```python
# AVERAGE SALARY OF USER WHO HAS PROGRAMING HOBBY
df4=df1[df1['ProgramHobby']=='Yes, I contribute to open source projects']
OS=Salary(df4)
```

```python
#PLOTTING BAR CHART FOR REPRESENTING THE DIFFERENCE IN SALARY.
objects=('No Hobby','Yes Hobby','Contribute to Open Source')
y_pos = np.arange(len(objects))
Mean = [NO,YES,OS]
plt.bar(y_pos, Mean, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Average Salary')
plt.show()
```

### CONCLUSION :
From the bar chart , it seems that user who contribute to open source has a little more salary than other users but the differece is not too much.
